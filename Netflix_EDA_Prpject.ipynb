```python
# Content type distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='type', palette='coolwarm')
plt.title("Distribution of TV Shows & Movies")
plt.show()
```

```python
# Correlation matrix (only applicable to numerical columns)
plt.figure(figsize=(8, 6))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation between Features")
plt.show()
```

```python
most_watched = df.sort_values(by='show_id', ascending=False).head(10)
most_watched
```

```python
# Most common ratings
plt.figure(figsize=(10, 4))
sns.countplot(data=df, y='rating', order=df['rating'].value_counts().index, palette='viridis')
plt.title("Distribution of Ratings")
plt.show()
```

```python
highest_rated = df.groupby(['type', 'rating']).size().unstack()
highest_rated.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title("Comparison of TV Shows & Movies by Rating")
plt.show()
```

```python
# Assuming 'date_added' is in a suitable format, handling potential leading/trailing spaces
df['release_month'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce').dt.month
monthly_releases = df.groupby('release_month')['title'].count()
monthly_releases
```

```python
# Best month for releasing content
df['release_month'] = pd.to_datetime(df['date_added'], format='mixed', errors='coerce').dt.month
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='release_month', palette='Blues')
plt.title("Best Month for Releasing Content")
plt.show()
```

```python
# Group by 'genre' and sum the 'view_count'
genre_views = df.groupby('listed_in')['show_id'].count()
genre_views
```

```python
# Most watched genres (Top 10)
genres = df['listed_in'].str.split(', ').explode()
genre_counts = genres.value_counts().head(10)
plt.figure(figsize=(10, 4))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='magma')
plt.xticks(rotation=45)
plt.title("Top 10 Most Watched Genres")
plt.show()
```

```python
# Group by 'release_year' and count the movies
yearly_releases = df.groupby('release_year')['title'].count()
yearly_releases
```

```python
# Movies released over the years
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='release_year', palette='coolwarm', order=sorted(df['release_year'].dropna().astype(int).unique()))
plt.xticks(rotation=45)
plt.title("Movies Released Over the Years")
plt.show()
```

```python
# Show ID & Director for 'House of Cards'
house_of_cards = df[df['title'] == 'House of Cards'][['show_id', 'director']]
print("House of Cards Details:")
house_of_cards
```

```python
# Movies released in the year 2000
year_2000_movies = df[(df['release_year'] == 2000) & (df['type'] == 'Movie')]
print("Movies Released in 2000:")
year_2000_movies[['title']]
```

```python
# TV Shows released in India
tv_shows_india = df[(df['type'] == 'TV Show') & (df['country'] == 'India')]
print("TV Shows Released in India:")
tv_shows_india[['title']]
```

```python
top_directors = df['director'].value_counts().head(10)
plt.figure(figsize=(10, 4))
sns.barplot(x=top_directors.index, y=top_directors.values, palette='rocket')
plt.xticks(rotation=45)
plt.title("Top 10 Directors on Netflix")
plt.show()
```

```python
# Count of movies/TV shows featuring Tom Cruise
tom_cruise_movies = df[df['cast'].str.contains('Tom Cruise', na=False)]
print("Movies/TV Shows Featuring Tom Cruise:")
tom_cruise_movies[['title', 'type']]
```

```python
# Movies rated 'TV-14' in Canada
tv_14_canada = df[(df['rating'] == 'TV-14') & (df['country'] == 'Canada')]
print("Movies with 'TV-14' Rating in Canada:")
tv_14_canada[['title']]
```
